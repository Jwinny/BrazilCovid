```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/wcota/covid19br/master/cases-brazil-states.csv'
dataframe = pd.read_csv(url)
```

```python
dataframe.head()
```

```python
valores_faltantes = dataframe.isnull().sum()
valores_faltantes
```

```python
total_celulas = np.product(dataframe.shape)
print(f"Quantidade de células no dataframe {total_celulas}")
total_sem_dados = valores_faltantes.sum()
print(f"\nQuantidade de células não preenchidas {total_sem_dados}")
porcentagem_ausencia_dados = (total_sem_dados/total_celulas)*100
print(f"\nPorcentagem de dados ausentes foi de: {porcentagem_ausencia_dados:.2f}%")
```

```python
dataframe_prenchido_com_zeros = dataframe.fillna(0)
dataframe_prenchido_com_zeros
```

```python
media_casos = dataframe_prenchido_com_zeros.groupby('state')['totalCases'].mean().sort_values(ascending=False)
media_casos = media_casos.drop('TOTAL')
```

```python
plt.figure(figsize=(12,6))
media_casos.plot(kind='bar', title='Média de Casos COVID-19 por Estado')
plt.ylabel('Casos Médios')
plt.xlabel('Estados')
plt.savefig('grafico_casos_por_estado.png')
plt.show()
```
